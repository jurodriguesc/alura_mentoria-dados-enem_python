#Análise da participação no ENEM 2021 por estado de origem dos participantes

```python
dataset_enem['SG_UF_PROVA']
```

```python
dataset_enem['SG_UF_PROVA'].unique()
```

```python
dataset_enem['SG_UF_PROVA'].value_counts()
```

```python
labels = dataset_enem['SG_UF_PROVA'].value_counts().index
values = dataset_enem['SG_UF_PROVA'].value_counts().values
```

```python
fig, ax = plt.subplots(figsize=(12, 5))
ax.bar(labels, values)
ax.set_title('Quantidade de participantes do ENEM 2021 por estado', fontsize=16)
ax.set_ylabel("Número de participantes")
ax.set_xlabel("Estados")
plt.show()
```
